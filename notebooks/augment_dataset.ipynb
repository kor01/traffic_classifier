```python
import tensorflow as tf
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.gridspec as gridspec
from tf_traffic_net.dataset import load_traffic_dataset
%matplotlib inline
```

```python
dataset = load_traffic_dataset('./traffic_data')
```

```python
images = dataset.train['features']
start = 100
figure = plt.figure(figsize=(16, 16))
gs = gridspec.GridSpec(
        nrows=2, ncols=5, left=0.1, bottom=0.25, right=0.95, top=0.95,
        wspace=0.05, hspace=0., width_ratios=[1] * 5)
for i in range(10):
    ax = plt.subplot(gs[i])
    ax.imshow(images[start + i])
plt.show()
```

```python
image = tf.placeholder('float32', shape=(32, 32, 3))
sess = tf.Session()
```

```python
brightness = tf.image.random_brightness(image, max_delta=1.0)
```

```python
brightness_images = []
for i in range(10):
    brightness_images.append(sess.run(brightness, feed_dict={image: images[start + i]}))
igure = plt.figure(figsize=(16, 16))
gs = gridspec.GridSpec(
        nrows=2, ncols=5, left=0.1, bottom=0.25, right=0.95, top=0.95,
        wspace=0.05, hspace=0., width_ratios=[1] * 5)
for i in range(10):
    ax = plt.subplot(gs[i])
    ax.imshow(brightness_images[i].astype('uint8'))
plt.show()
```

```python
contrast = tf.image.random_contrast(image, 0.2, 1.0)
```

```python
contrast_images = []
for i in range(10):
    contrast_images.append(sess.run(contrast, feed_dict={image: images[start + i]}))
igure = plt.figure(figsize=(16, 16))
gs = gridspec.GridSpec(
        nrows=2, ncols=5, left=0.1, bottom=0.25, right=0.95, top=0.95,
        wspace=0.05, hspace=0., width_ratios=[1] * 5)
for i in range(10):
    ax = plt.subplot(gs[i])
    ax.imshow(contrast_images[i].astype('uint8'))
plt.show()
```

```python
saturation = tf.image.random_saturation(image, lower=1.0, upper=3.0)
```

```python
sat_images = []
for i in range(10):
    sat_images.append(sess.run(saturation, feed_dict={image: images[start + i]}))
igure = plt.figure(figsize=(16, 16))
gs = gridspec.GridSpec(
        nrows=2, ncols=5, left=0.1, bottom=0.25, right=0.95, top=0.95,
        wspace=0.05, hspace=0., width_ratios=[1] * 5)
for i in range(10):
    ax = plt.subplot(gs[i])
    ax.imshow(sat_images[i].astype('uint8'))
plt.show()
```

```python
batch_size = 10
images_ph = tf.placeholder(dtype='float32', shape=(batch_size, 32, 32, 3))
angles = tf.random_uniform((batch_size,), -np.pi / 6, np.pi / 6)
rotation = tf.contrib.image.rotate(images_ph, angles)
```

```python
rotated_images = sess.run(rotation, {images_ph: images[start:start+10]})
```

```python
figure = plt.figure(figsize=(16, 16))
gs = gridspec.GridSpec(
        nrows=2, ncols=5, left=0.1, bottom=0.25, right=0.95, top=0.95,
        wspace=0.05, hspace=0., width_ratios=[1] * 5)
for i in range(10):
    ax = plt.subplot(gs[i])
    ax.imshow(rotated_images[i].astype('uint8'))
plt.show()
```

```python
hue = tf.squeeze(tf.image.rgb_to_grayscale(
    tf.image.random_hue(image, max_delta=0.1)))
```

```python
hue_images = []
for i in range(10):
    hue_images.append(sess.run(hue, feed_dict={image: images[start + i]}))
igure = plt.figure(figsize=(16, 16))
gs = gridspec.GridSpec(
        nrows=2, ncols=5, left=0.1, bottom=0.25, right=0.95, top=0.95,
        wspace=0.05, hspace=0., width_ratios=[1] * 5)
for i in range(10):
    ax = plt.subplot(gs[i])
    ax.imshow(hue_images[i].astype('uint8'))
plt.show()
```
